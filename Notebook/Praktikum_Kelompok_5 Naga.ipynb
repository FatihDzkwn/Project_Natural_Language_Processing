*   Case Folding

```python
def clean_ulasan(data_shopee, Ulasan):

    # Ubah data menjadi huruf kecil
    data_shopee['Ulasan_clean'] = data_shopee[Ulasan].str.lower()

    # Hapus spasi ganda atau baris kosong
    data_shopee['Ulasan_clean'] = data_shopee['Ulasan_clean'].str.strip()

    # Hapus tanda baca
    data_shopee['Ulasan_clean'] = data_shopee['Ulasan_clean'].str.replace('[^\w\s]', '', regex=True)

    # Hapus huruf yang berulang
    data_shopee['Ulasan_clean'] = data_shopee['Ulasan_clean'].str.replace(r'(.)\1{2,}', r'\1', regex=True)

    return data_shopee
```

```python
# Jalankan fungsi
data_shopee_clean = clean_ulasan(data_shopee, 'Ulasan')

print(data_shopee_clean[['Ulasan', 'Ulasan_clean']])
```

* Load Slang

```python
#import library yang dibutuhkan
import re
import json
import requests
```

```python
# crawl data slang
slang_url = "https://raw.githubusercontent.com/louisowen6/NLP_bahasa_resources/master/combined_slang_words.txt"
response = requests.get(slang_url)
slang_dict = json.loads(response.text)
print(slang_dict)
```

*   Stemming

```python
#install sastrawi untuk bahasa indonesia
!pip install Sastrawi
```

```python
#import library sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
#using sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Cleaning Text dengan regex Library
def clean_text(text):

    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    words = text.split()

    # normalisasi slang
    words = [slang_dict[word] if word in slang_dict else word for word in words]


    text = " ".join(words)

    # stemming
    text = stemmer.stem(text)

    return text
```

```python
# Jalankan Pipeline
data_shopee_clean['clean_ulasan'] = data_shopee_clean['Ulasan'].apply(clean_text)

print(data_shopee_clean[['Ulasan', 'clean_ulasan']])
```

* Tokenize

```python
!pip install spacy
import spacy

nlp = spacy.blank("id")  # tokenizer saja

def spacy_tokenize(text):
    doc = nlp(text)
    return [token.text for token in doc]

data_shopee_clean['tokens'] = data_shopee_clean['clean_ulasan'].apply(spacy_tokenize)
```
