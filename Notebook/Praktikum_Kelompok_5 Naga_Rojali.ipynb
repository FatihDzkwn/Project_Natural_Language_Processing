# **2. Text Preprocessing**

*   Case Folding

```python
def clean_ulasan(data_shopee, Ulasan):

    # Ubah data menjadi huruf kecil
    data_shopee['Ulasan_clean'] = data_shopee[Ulasan].str.lower()

    # Hapus spasi ganda atau baris kosong
    data_shopee['Ulasan_clean'] = data_shopee['Ulasan_clean'].str.strip()

    # Hapus tanda baca
    data_shopee['Ulasan_clean'] = data_shopee['Ulasan_clean'].str.replace('[^\w\s]', '', regex=True)

    # Hapus huruf yang berulang
    data_shopee['Ulasan_clean'] = data_shopee['Ulasan_clean'].str.replace(r'(.)\1{2,}', r'\1', regex=True)

    return data_shopee
```

```python
# Jalankan fungsi
data_shopee_clean = clean_ulasan(data_shopee, 'Ulasan')

print(data_shopee_clean[['Ulasan', 'Ulasan_clean']])
```

*   Stop Word Removal

```python
#Load stop word
stopwords_url = "https://raw.githubusercontent.com/datascienceid/stopwords-bahasa-indonesia/master/stopwords_id_satya.txt"
stopwords = pd.read_csv(stopwords_url, header=None)[0].tolist()
stopwords
```

```python
# Fungsi untuk menghapus stopwords dari teks
def remove_stopwords(text):
    # Memisahkan kalimat menjadi kata-kata, lalu membuang kata yang ada di daftar stopwords
    return ' '.join([word for word in str(text).split() if word not in stopwords])

# Terapkan ke kolom Ulasan_clean
data_shopee_clean['Ulasan_clean'] = data_shopee_clean['Ulasan_clean'].apply(remove_stopwords)

# Tampilkan perbandingannya
print("Hasil setelah penghapusan Stopwords:")
print(data_shopee_clean[['Ulasan', 'Ulasan_clean']].head())
```

* Load Slang

```python
#import library yang dibutuhkan
import re
import json
import requests
```

```python
# crawl data slang
slang_url = "https://raw.githubusercontent.com/louisowen6/NLP_bahasa_resources/master/combined_slang_words.txt"
response = requests.get(slang_url)
slang_dict = json.loads(response.text)
print(slang_dict)
```

```python
# Fungsi untuk mengubah kata gaul (slang) menjadi kata baku
def replace_slang(text):
    # Memecah kalimat menjadi daftar kata
    words = str(text).split()
    # Mengganti kata jika ada di dalam kamus slang, jika tidak tetap gunakan kata aslinya
    words = [slang_dict[word] if word in slang_dict else word for word in words]
    # Menggabungkan kembali menjadi kalimat
    return ' '.join(words)

# Terapkan fungsi ke kolom ulasan yang sudah bersih
data_shopee_clean['Ulasan_clean'] = data_shopee_clean['Ulasan_clean'].apply(replace_slang)

# Tampilkan hasil perbandingannya
print("Hasil setelah perbaikan kata slang:")
display(data_shopee_clean[['Ulasan', 'Ulasan_clean']].head())
```

*   Stemming

```python
#install sastrawi untuk bahasa indonesia
!pip install Sastrawi
```

```python
#import library sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
#using sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Cleaning Text dengan regex Library
def clean_text(text):

    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    words = text.split()

    # normalisasi slang
    words = [slang_dict[word] if word in slang_dict else word for word in words]


    text = " ".join(words)

    # stemming
    text = stemmer.stem(text)

    return text
```

```python
# Jalankan Pipeline
data_shopee_clean['Ulasan_clean'] = data_shopee_clean['Ulasan_clean'].apply(clean_text)

print(data_shopee_clean[['Ulasan', 'Ulasan_clean']])
```

```python
# Impor library-nya
!pip install gensim
from gensim.models import Word2Vec

# Buat variabel 'sentences' dari data ulasan yang sudah bersih
# Pecah setiap kalimat menjadi daftar kata-kata
sentences = [str(text).split() for text in data_shopee_clean['Ulasan_clean']]

# TRAINING SKIP-GRAM & CBOW

EMBEDDING_DIM = 100   # Dimensi vektor
WINDOW        = 5     # Ukuran konteks window
MIN_COUNT     = 3     # Minimum frekuensi kata
EPOCHS        = 20    # Jumlah epoch training
WORKERS       = 4     # Jumlah thread CPU

print("Training Skip-Gram...")
model_sg = Word2Vec(
    sentences,
    vector_size = EMBEDDING_DIM,
    window      = WINDOW,
    min_count   = MIN_COUNT,
    sg          = 1,        # 1 = Skip-Gram
    epochs      = EPOCHS,
    workers     = WORKERS
)
print(f"✓ Skip-Gram selesai | Vocab: {len(model_sg.wv):,} kata")

print("\nTraining CBOW...")
model_cbow = Word2Vec(
    sentences,
    vector_size = EMBEDDING_DIM,
    window      = WINDOW,
    min_count   = MIN_COUNT,
    sg          = 0,        # 0 = CBOW
    epochs      = EPOCHS,
    workers     = WORKERS
)
print(f"✓ CBOW selesai | Vocab: {len(model_cbow.wv):,} kata")
```

```python
# USE CASE 1: MENCARI KATA YANG PALING MIRIP
# Aplikasi: Search engine, autocomplete, rekomendasi konten

def tampilkan_kata_mirip(model, kata, model_nama="Model", top_n=8):
    print(f"\n{'='*50}")
    print(f"[{model_nama}] Kata paling mirip dengan '{kata}':")
    print(f"{'='*50}")
    if kata not in model.wv:
        print(f"  Kata '{kata}' tidak ada dalam vocabulary!")
        return
    hasil = model.wv.most_similar(kata, topn=top_n)
    for rank, (w, score) in enumerate(hasil, 1):
        bar = "█" * int(score * 20)
        print(f"  {rank}. {w:<20} {score:.4f}  {bar}")

# Uji dengan beberapa kata
for kata_uji in ["kabel", "helm", "wajah", "bahan", "bagus", "cepat"]:
    tampilkan_kata_mirip(model_sg, kata_uji, "Skip-Gram")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# USE CASE 2: KEMIRIPAN KALIMAT (Sentence Similarity)
# Aplikasi: Deteksi plagiarisme, FAQ matching, QA system

def kalimat_ke_vektor(model, kalimat):
    """Rata-rata vektor semua kata dalam kalimat (mean pooling)"""
    kata_list = kalimat.lower().split() 
    vektors = [model.wv[k] for k in kata_list if k in model.wv]
    if not vektors:
        return np.zeros(model.vector_size)
    return np.mean(vektors, axis=0)

def hitung_kemiripan_kalimat(model, kalimat1, kalimat2):
    v1 = kalimat_ke_vektor(model, kalimat1).reshape(1, -1)
    v2 = kalimat_ke_vektor(model, kalimat2).reshape(1, -1)
    return cosine_similarity(v1, v2)[0][0]

# Dataset kalimat uji
kalimat_uji = [
    "Kabel terminal ini sangat aman dan lubangnya banyak",    # Elektronik (K1)
    "Colokan berfungsi dengan baik tidak ada konslet",         # Elektronik (K2)
    "Ciloknya enak kenyal dan bumbu kacangnya mantap",         # Makanan (K3)
    "Baso aci dan cilok kuahnya pedas bikin nagih",            # Makanan (K4)
    "Serum ini cepat meresap dan melembabkan wajah",     # Kecantikan (K5)
    "Kulit jadi halus dan glowing pakai pelembab ini",        # Kecantikan (K6)
    "Helm bogonya bagus kacanya bening nyaman dipakai",        # Otomotif (K7)
    "Busa helm empuk dan ukuran pas di kepala",                # Otomotif (K8)
]

print("MATRIKS KEMIRIPAN KALIMAT")
print("=" * 70)
n = len(kalimat_uji)
matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        matrix[i][j] = hitung_kemiripan_kalimat(model_sg, kalimat_uji[i], kalimat_uji[j])

# Tampilkan heatmap
fig, ax = plt.subplots(figsize=(10, 8))
label_pendek = [f"K{i+1}" for i in range(n)]
im = ax.imshow(matrix, cmap="Blues", vmin=0, vmax=1)
ax.set_xticks(range(n)); ax.set_yticks(range(n))
ax.set_xticklabels(label_pendek); ax.set_yticklabels(label_pendek)
plt.colorbar(im, ax=ax)
ax.set_title("Sentence Similarity Matrix (Skip-Gram)", fontsize=13)
for i in range(n):
    for j in range(n):
        ax.text(j, i, f"{matrix[i,j]:.2f}", ha="center", va="center", fontsize=9)
plt.tight_layout()
plt.show()

# Legenda kalimat
print("\nKeterangan:")
for i, k in enumerate(kalimat_uji):
    print(f"  K{i+1}: {k}")
```

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.cm as cm

# USE CASE 3: KLASTERISASI KATA
# Aplikasi: Topic modeling, pengelompokan dokumen otomatis

def klasterisasi_kata(model, n_clusters=4, top_words=200): # n_clusters diset 4
    # Ambil top N kata berdasarkan frekuensi
    kata_list  = [w for w in list(model.wv.key_to_index.keys())[:top_words]]
    vektor_list = np.array([model.wv[w] for w in kata_list])

    # K-Means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(vektor_list)

    # Kelompokkan hasil
    klaster = {i: [] for i in range(n_clusters)}
    for kata, label in zip(kata_list, labels):
        klaster[label].append(kata)

    return klaster, vektor_list, labels, kata_list

# Eksekusi klasterisasi
n_klaster_tujuan = 4
klaster, vektors, labels, kata_list = klasterisasi_kata(model_cbow, n_clusters=n_klaster_tujuan)

print("HASIL KLASTERISASI KATA")
print("=" * 60)
for i, kata_dalam_klaster in klaster.items():
    print(f"\nKlaster {i+1} ({len(kata_dalam_klaster)} kata):")
    # Menampilkan 15 kata pertama di setiap klaster
    print("  " + ", ".join(kata_dalam_klaster[:15]) + "...")

# Visualisasi dengan t-SNE
print("\nMembuat visualisasi t-SNE...")
tsne   = TSNE(n_components=2, random_state=42, perplexity=30)
coords = tsne.fit_transform(vektors)

# Gunakan Colormap Biru (karena kamu ingin tema biru)
# Kita gunakan 'Blues' atau 'winter' agar senada
colors = plt.cm.winter(np.linspace(0, 1, n_klaster_tujuan)) 

plt.figure(figsize=(12, 9))
for i in range(n_klaster_tujuan):
    mask = labels == i
    plt.scatter(coords[mask, 0], coords[mask, 1],
                color=colors[i], label=f"Klaster {i+1}", alpha=0.7, s=80, edgecolors='white')

# Label hanya beberapa kata agar tidak menumpuk
for idx, kata in enumerate(kata_list[:100]):   
    plt.annotate(kata, coords[idx], fontsize=8, alpha=0.8)

plt.title("Visualisasi Klaster Kata Produk Shopee (t-SNE)", fontsize=14, fontweight='bold')
plt.legend(loc="upper right")
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
    # USE CASE 4: ANALOGI BELANJA SHOPEE
    # Aplikasi: Knowledge graph, chatbot, NLP reasoning

def word_analogy(model, positif, negatif, top_n=5):
        print(f"\n{'='*55}")
        # Format tampilan agar lebih rapi
        rumus = f"{' + '.join(positif)} - {' + '.join(negatif)}"
        print(f"  Analogi: {rumus} = ?")
        print(f"{'='*55}")
        try:
            # Gunakan model yang sedang aktif (bisa model_sg atau model_cbow)
            hasil = model.wv.most_similar(positive=positif, negative=negatif, topn=top_n)
            for w, score in hasil:
                print(f"  → {w:<20} (skor: {score:.4f})")
        except KeyError as e:
            print(f"  Kata tidak ditemukan dalam dataset: {e}")

   # --- CONTOH ANALOGI KONSEP ---

# 1. Hubungan Antar Kategori
# "Kabel Terminal ke Elektronik itu seperti Helm ke...?"
word_analogy(model_sg, ["helm", "elektronik"], ["kabel"])

# 2. Hubungan Produk ke Sifat/Hasil
# "Cilok itu enak, kalau Serum itu...?"
word_analogy(model_sg, ["serum", "enak"], ["cilok"])

# 3. Hubungan Kualitas & Masalah (Analogi Negatif)
# "Ori itu lawan dari Jelek, kalau Bagus lawan dari...?"
word_analogy(model_sg, ["jelek", "original"], ["ori"])

# 4. Hubungan Terminologi (Slang/Typo Shopee)
# "Original disingkat Ori, kalau Rekomendasi disingkat...?"
word_analogy(model_sg, ["rekomendasi", "ori"], ["original"])
```

* Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Inisialisasi TfidfVectorizer
tfidf_vec = TfidfVectorizer(max_features=5000)

# Fit dan Transform data
X_train_tfidf_matrix = tfidf_vec.fit_transform(X_train_raw)
X_test_tfidf_matrix = tfidf_vec.transform(X_test_raw)

# Konversi ke DataFrame untuk Visualisasi (Sesuai gaya dosen)
X_train_tfidf_df = pd.DataFrame(
    X_train_tfidf_matrix.toarray(), 
    columns=tfidf_vec.get_feature_names_out()
)

print("Tampilan DataFrame TF-IDF (Latihan):")
X_train_tfidf_df.head()
```

* Word Embedding

* Word2Vec

```python
# Fungsi untuk mengubah ulasan menjadi rata-rata vektor (Mean Pooling)
def ulasan_ke_vektor(teks, model):
    words = str(teks).split()
    # Ambil vektor jika kata ada di kamus Word2Vec
    vectors = [model.wv[w] for w in words if w in model.wv]
    if not vectors:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)

# Ekstraksi fitur Word2Vec untuk Training dan Testing
X_train_w2v = np.array([ulasan_ke_vektor(s, model_sg) for s in X_train_raw])
X_test_w2v = np.array([ulasan_ke_vektor(s, model_sg) for s in X_test_raw])

# Konversi ke DataFrame agar seragam dengan TF-IDF
X_train_w2v_df = pd.DataFrame(X_train_w2v)
print("\nTampilan DataFrame Word2Vec (Latihan):")
X_train_w2v_df.head()
```

```python
!pip install seaborn
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Inisialisasi dan Training
nb_tfidf = MultinomialNB()
nb_tfidf.fit(X_train_tfidf_matrix, y_train)

# Prediksi
y_pred_nb_tfidf = nb_tfidf.predict(X_test_tfidf_matrix)

# Evaluasi
print("--- Naive Bayes (TF-IDF) ---")
print("Accuracy:", accuracy_score(y_test, y_pred_nb_tfidf))
print(classification_report(y_test, y_pred_nb_tfidf))

# Confusion Matrix
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_test, y_pred_nb_tfidf), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix: NB + TF-IDF")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.svm import SVC

# Inisialisasi dan Training
svm_tfidf = SVC(kernel='linear', probability=True)
svm_tfidf.fit(X_train_tfidf_matrix, y_train)

# Prediksi
y_pred_svm_tfidf = svm_tfidf.predict(X_test_tfidf_matrix)

# Evaluasi
print("--- SVM (TF-IDF) ---")
print("Accuracy:", accuracy_score(y_test, y_pred_svm_tfidf))
print(classification_report(y_test, y_pred_svm_tfidf))

# Confusion Matrix
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_test, y_pred_svm_tfidf), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix: SVM + TF-IDF")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
def evaluate_model(name, y_test, y_pred):
    report = classification_report(y_test, y_pred, output_dict=True)
    return {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision (weighted)": report['weighted avg']['precision'],
        "Recall (weighted)": report['weighted avg']['recall'],
        "F1-score (weighted)": report['weighted avg']['f1-score']
    }

results = []

# NB + TF-IDF
results.append(evaluate_model("NB + TF-IDF", y_test, y_pred_nb_tfidf))

# LogReg + Word2Vec
results.append(evaluate_model("LogReg + Word2Vec", y_test, y_pred_lr_w2v))

# SVM + TF-IDF
results.append(evaluate_model("SVM + TF-IDF", y_test, y_pred_svm_tfidf))

# Tampilkan Tabel
results_df = pd.DataFrame(results)
print("\n--- FINAL COMPARISON TABLE ---")
print(results_df)
```

```python
!pip install gradio
import re
import gradio as gr
import numpy as np

# Fungsi Pembersihan Teks Tunggal - SESUAIKAN DENGAN TRAINING PIPELINE
def clean_text_single(text):
    # Sama dengan fungsi clean_text() yang digunakan saat training
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # Hapus tanda baca
    text = re.sub(r'(.)\1{2,}', r'\1', text)  # Hapus karakter berulang
    words = text.split()
    
    # Normalisasi slang (PENTING!)
    words = [slang_dict[word] if word in slang_dict else word for word in words]
    
    text = " ".join(words)
    
    # Stemming dengan Sastrawi (PENTING!)
    text = stemmer.stem(text)
    
    return text

# Fungsi Prediksi untuk Gradio
def predict_review(text):
    cleaned = clean_text_single(text)
    print(f"Teks asli: {text}")
    print(f"Teks bersih: {cleaned}")
    
    vectorized = tfidf_vec.transform([cleaned])
    pred = svm_tfidf.predict(vectorized)[0]
    prob = svm_tfidf.predict_proba(vectorized)[0]
    
    # Return dalam format dict untuk Gradio Label
    # Gunakan label kelas sebagai key, probability sebagai value
    classes = svm_tfidf.classes_
    result = {classes[i]: float(prob[i]) for i in range(len(classes))}
    
    return result
```
